```python
import sqlite3
from sqlite3 import Error

import pandas as pd
```

```python
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)

    return conn
```

```python
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()

    query1 = """
        SELECT *
        FROM FACILITIES
        """
    cur.execute(query1)

    rows = cur.fetchall()

    for row in rows:
        print(row)
```

```python
def main():
    database = "sqlite_db_pythonsqlite.db"

    # create a database connection
    conn = create_connection(database)
    with conn:
        print("2. Query all tasks")
        select_all_tasks(conn)
```

```python
if __name__ == '__main__':
    main()
```

```python
# Q10: Produce a list of facilities with a total revenue less than 1000.
# The output of facility name and total revenue, sorted by revenue. Remember
# that there's a different cost for guests and members! 

database = "sqlite_db_pythonsqlite.db" 
conn = create_connection(database)

pd.read_sql_query('SELECT facility_name AS "Facility Name", SUM(Cost) AS "Total_Revenue" \
                    FROM (SELECT Facilities.name AS facility_name, CASE WHEN bookings.memid != 0 THEN Bookings.slots*Facilities.membercost \
                    WHEN Bookings.memid = 0 THEN Bookings.slots*Facilities.guestcost END as Cost FROM Members, Facilities, Bookings \
                    WHERE Facilities.facid = Bookings.facid and Bookings.memid = Members.memid) AS Subquery \
                    GROUP BY facility_name HAVING SUM(Cost) < 1000 ORDER BY Total_Revenue', conn)

```

```python
# Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order 

pd.read_sql_query('SELECT DISTINCT Members.surname || ", " || Members.firstname AS "Member",  \
                   Recommender.surname || ", " || Recommender.firstname AS "Recommended By" \
                   FROM Members INNER JOIN Members AS Recommender ON Members.recommendedby = Recommender.memid \
                   WHERE Members.memid != 0 ORDER BY Members.surname, Members.firstname', conn)
```

```python
# Q12: Find the facilities with their usage by member, but not guests

pd.read_sql_query('SELECT Facilities.name as "Facility Name", SUM(Bookings.slots) AS "Total Usage by Members" \
                   FROM Bookings, Facilities WHERE Bookings.memid != 0 AND Bookings.facid = Facilities.facid \
                   GROUP BY "Facility Name"', conn)
```

```python
# Q13: Find the facilities usage by month, but not guests

pd.read_sql_query('SELECT strftime("%m", Starttime) AS "Month", Facilities.name AS "Facility Name", SUM(Bookings.slots) AS "Total Usage" \
                   FROM Bookings, Facilities WHERE Bookings.memid != 0 and Bookings.facid = Facilities.facid \
                   GROUP BY Month, "Facility Name"', conn)

```
